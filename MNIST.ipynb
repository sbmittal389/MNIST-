```python
def distance(p1,p2):
    if np.sqrt((sum(p2-p1))**2)==0:
        return 1,np.sqrt((sum(p2-p1))**2)
    else:
        return 0,0

def knn(x,y,test_fruit, k=10):
    d=[]
    m=x.shape[0]
    for i in range(m):
        a,b=distance(test_fruit,x[i])
        if(a==1):
            dist=b
            d.append((dist,y[i]))
    d=np.array(sorted(d))
    d=d[:k]
    t=np.unique(d[:,1],return_counts=True)
    idx=t[1].argmax()
    pred=int(t[0][idx])
    return(pred)
```

```python
for i in range(10):
    pred=knn(x,y,x[i])
    print(pred)
```
